```python
tang_nifh = pd.read_csv("../csv/nifh_di/Tang_GeophysicalResLett_2019_Data_depth_int.csv")
print(tang_nifh.info())
tang_nifh.head()
```

```python
tang_cols = ['Trichodesmium nifH Gene (x106 copies m-2)','UCYN-A nifH Gene (x106 copies m-2)','UCYN-B nifH Gene (x106 copies m-2)']

setND(tang_nifh, tang_cols)
#first step
objToNum(tang_nifh)
#tang_nifh.fillna(0, inplace=True)

#second step
tang_nifh_df = tang_nifh
#tang_nifh_df = removeEmptyRows(tang_cols,tang_nifh_df)

#check that nothign is lost
print(len(tang_nifh_df))
```

```python
from parula_diy import parula_map

plotColsOnMap(
    tang_cols,tang_nifh_df,
    constant=(10**6), 
    transf=(lambda x: np.log10(x+10)),
    min_lim=6,
    max_lim=11,
    cmap=parula_map,
    colorbar_label=" log10 ( 10^6 * nifH Gene (copies m-2))",
    title="Tang depth integrated data presented in the exel file",
    s=30
    )
```

## Raw data vs predictions

Here I wanted to visualize the comparison of raw data vs the final predictions made.

```python
raw_data = pd.read_csv("../csv/datasets/nifh_dataset.csv")
predictions = pd.read_csv("../csv/predictions/pred_sep.csv")
```

```python
plotColsOnMap(
    y_columns, 
    raw_data, 
    constant=(10**3), 
    transf=(lambda x: np.log10(x+1)),
    cmap=parula_map,
    specific_depth=5,
    s=15,
    colorbar_label="nifH Gene (x10^3 copies L^-1)",
    title="Raw data in L at 5m"
    )
```

```python
plotColsOnMap(
    y_columns, 
    predictions, 
    constant=(10**3), 
    transf=(lambda x: np.log10(x+1)),
    cmap=parula_map,
    min_lim=0, 
    max_lim=5,
    specific_depth=5,
    colorbar_label="nifH Gene (x10^3 copies L^-1)",
    title="Predictions at 5m in L",
    color_steps=10
)
```

```python
plotColsOnMap(
    cols=y_columns, 
    df=predictions, 
    overlay=raw_data,
    constant=(10**3), 
    transf=(lambda x: np.log10(x+1)),
    min_lim=0, 
    max_lim=5,
    cmap=parula_map,
    specific_depth=5,
    s=7,
    colorbar_label="nifH Gene (x10^3 copies L^-1)",
    title="Predictions and raw data at 5m in L. Comparison",
    color_steps=10
)
```

```python
plotColsOnMap(
    cols=y_columns, 
    df=predictions, 
    overlay=raw_data,
    constant=(10**3), 
    transf=(lambda x: np.log10(x+1)),
    cmap=parula_map,
    s=7,
    colorbar_label="nifH Gene (x10^3 copies L^-1)",
    title="Predictions and raw data avg over depth in L. Comparison"
)
```
